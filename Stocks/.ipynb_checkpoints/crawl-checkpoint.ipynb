```python
import baostock as bs
import pandas as pd
```

```python
def get_stock(stock_num, stock_type, start, end):
    stock = stock_num+'.'+stock_type
    lg = bs.login()
    rs = bs.query_history_k_data(stock,"date,open,high,low,close,volume,amount", 
                                 start_date = start, end_date = end, frequency="d")
    #print('query_history_k_data respond error_code:'+rs.error_code)
    #print('query_history_k_data respond  error_msg:'+rs.error_msg)
    data_list = []
    while (rs.error_code == '0') & rs.next(): 
        data_list.append(rs.get_row_data())
    result = pd.DataFrame(data_list, columns=rs.fields)
    df = result
    if len(df) == 0:
        print("No such stock!")
        return 0
    df.columns = ['交易日期','开盘价','最高价','最低价','收盘价','成交数量(股)','成交金额(元)']
    df.to_csv(stock+'.csv',index = False)
    return df
```

```python
def get_stock_to_excel(stock_num, stock_type, start, end):
    stock = stock_num+'.'+stock_type
    lg = bs.login()
    rs = bs.query_history_k_data(stock,"date,open,high,low,close,volume,amount", 
                                 start_date = start, end_date = end, frequency="d")
    print('query_history_k_data respond error_code:'+rs.error_code)
    print('query_history_k_data respond  error_msg:'+rs.error_msg)
    data_list = []
    while (rs.error_code == '0') & rs.next(): 
        data_list.append(rs.get_row_data())
    result = pd.DataFrame(data_list, columns=rs.fields)
    df = result
    df.columns = ['交易日期','开盘价','最高价','最低价','收盘价','成交数量(股)','成交金额(元)']
    df.to_excel(stock+'.xlsx',index = False)
    return df
```

```python
stock_num = 600000
SUM = 0
for i in range(0,9999):
    
    if SUM == 50:
        break
    stock_num += 1
    try:
        get_stock(str(stock_num),'SH','2020-07-22','2021-07-21')
        print(stock_num)
        print(SUM)
        try:
            pd.read_csv(str(stock_num)+'.SH.csv')
            SUM += 1
        except:
            continue
    except:
        continue
```
