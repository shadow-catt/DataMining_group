```python
# separate the dataframe
def separate(df, proportion=.8):
    rows=df.shape[0]
    p_train = int(np.floor(rows * proportion))
    train = df.iloc[:p_train]
    test = df.iloc[p_train:]
    return train,test

# mix the order of dataframe
def mix_df_order(df):
    rows=df.shape[0]
    col = df.columns.values.tolist()
    # mix the number sequence of rows
    index_list=list(range(rows))
    np.random.shuffle(index_list)
    
    # store the random line
    mix_df = df.iloc[index_list]
    return mix_df

# random get the traning set and testing set
def mix_and_sep(df, proportion=.8):
    mix=mix_df_order(df)
    return separate(mix, proportion)
```
